```python
import csv 
  
# opening the CSV file 
with open('expanded-URLs.csv', mode ='r')as file: 
    
    # reading the CSV file 
    csvFile = csv.reader(file)
    next(csvFile)
    
    with open('final-URLs.csv', mode ='w')as write_file: 
        csvFileWriter = csv.writer(write_file)
        
    
    # displaying the contents of the CSV file 
        for lines in csvFile:
            try:
                r = requests.get(lines[0])
                
                r.raise_for_status()
                csvFileWriter.writerow([lines[0], lines[1], r.url, r.status_code]) 
                
            except requests.exceptions.HTTPError as err:
                #raise SystemExit(err)
                #print(lines[0])
                csvFileWriter.writerow([lines[0], lines[1], lines[0], r.status_code])
```

```python
import csv 
  
# opening the CSV file 
with open('expanded-URLs-Copy1.csv', mode ='r')as file: 
    
    # reading the CSV file 
    csvFile = csv.reader(file)
    next(csvFile)
    
    with open('final-URLs.csv', mode ='w')as write_file: 
        csvFileWriter = csv.writer(write_file)
        
    
    # displaying the contents of the CSV file 
        for lines in csvFile:
            try:
                session = requests.Session()
                retry = Retry(connect=3, backoff_factor=0.5)
                adapter = HTTPAdapter(max_retries=retry)
                session.mount('http://', adapter)
                session.mount('https://', adapter)

                # get the urls of web page and save to file
                # if ok and pass if not
                #r = requests.get(lines[0], verify=False)
                r = session.head(lines[0])
                
                r.raise_for_status()
                if r.ok:
                    csvFileWriter.writerow([lines[0], lines[1], r.url, r.status_code]) 
                    
                else:
                    continue
                
            except requests.exceptions.HTTPError as err:
                #raise SystemExit(err)
                #print(lines[0])
                csvFileWriter.writerow([lines[0], lines[1], lines[0], r.status_code])
            except requests.exceptions.ConnectionError as err1:
                csvFileWriter.writerow([lines[0], lines[1], lines[0], r.status_code])
```
